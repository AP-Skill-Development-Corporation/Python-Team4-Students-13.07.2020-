```python
## 1. Default Arguments
def calculations(a,b=5):
    x=a+b
    print("addition",x)
    y=a-b
    print("sub",y)
    return (x,y)
calculations(20)    
```

```python
def calculations(a,b=5):
    x=a+b
    print("addition",x)
    y=a-b
    print("sub",y)
    return (x,y)
calculations(3)   ## default arguements   # b=2
```

```python
## 2.Positional Arguments
def addition(x,y):
    s=x+y
    return s
addition(20,60)   # x=20 y= 60
```

```python
addition(60,20) #x=60 y=20
```

```python
##3. Keyword Arguments
def funn(a,b):
    a=9
    b=10
    print(a,b)
funn(a=45,b=8)
```

```python
funn(a=4,b=7)
```

```python
# 4.arbitry Arguments /variable length(*)
def varfun(*n):
    for i in n:
        print(i)
varfun("jntua","ace","bvc","svew","git")
```

```python
def varfun(*n):
    for i in n:
        print(i,end="\t")
varfun("jntua","ace","bvc","svew","git")
```

## Recursive  Function
- A function will be called itself  with in the function defination

```python
def function1():
    print("python")
    function1()
function1()   
```

```python
# factorial 
#5!=5*4*3*2*1(number-1)
def fact(n):
    p=1
    for i in  range(1,n+1):
        print(i) # 1 2 3 4 5
        p=p*i #1*2*3*4*5
    return p
fact(5)   
```

```python
def factioral(n):
    if n==1 or n==0:
        return 1
    else:
         return n*factioral(n-1)  # n==5=5*fact(5-1)
                                 # 4*fact(4-1)
                                 # 3*2*1
        
        
fact(5)
```
